````python
# Data Structures and Algorithms Homework

#**Student Name:** Stephen Waita  
#**Course:** Data Structures and Algorithms  
#**Date:** 9/13/2024  

## Introduction

#This notebook presents the solutions to the homework on analyzing the time complexity of algorithms, including modifications to improve their running time and comparisons against other common complexity functions.

## Part 1: Analyzing the Algorithm

### Original Algorithm

#```python
def algorithm(a, b):
    for i in range(1, a + 1):
        for j in range(1, b + 1):
            print("(i,j) =(", i, ",", j, ")")
            if i == j and a % i == 0 and b % j == 0:
                print(i, " is a common divisor of ", a, " and ", b)

def improved_algorithm(a, b):
    n = min(a, b)
    for i in range(1, n + 1):
        if a % i == 0 and b % i == 0:
            print(i, "is a common divisor of", a, "and", b)

print("Original Algorithm Output:")
print("\nFor (a, b) = (4, 20):")
algorithm(4, 20)
print("\nFor (a, b) = (5, 30):")
algorithm(5, 30)
print("\nFor (a, b) = (40, 100):")
algorithm(40, 100)

print("\nImproved Algorithm Output:")
print("\nFor (a, b) = (4, 20):")
improved_algorithm(4, 20)
print("\nFor (a, b) = (5, 30):")
improved_algorithm(5, 30)
print("\nFor (a, b) = (40, 100):")
improved_algorithm(40, 100)

import numpy as np
import matplotlib.pyplot as plt
import math  # Import the math module

# Define the range of n values
n_values = np.arange(1, 10, 0.1)

# Define the functions
f_log = np.log(n_values)
f_n = n_values
f_n2 = n_values ** 2
f_2n = 2 ** n_values
f_fact = np.array([math.factorial(int(n)) for n in np.arange(1, 10)])  # Factorial grows quickly, so limit to integers

# Plot the functions
plt.figure(figsize=(10, 6))

plt.plot(n_values, f_n, label="Improved Algorithm: O(n)", color="blue")
plt.plot(n_values, f_log, label="f(n) = log(n)", color="green")
plt.plot(n_values, f_n2, label="f(n) = n^2", color="orange")
plt.plot(n_values, f_2n, label="f(n) = 2^n", color="red")
plt.plot(np.arange(1, 10), f_fact, label="f(n) = n!", color="purple")  # Factorial only on integer n

# Remove y-axis limits to see the full plot
plt.yscale('log')  # Use a logarithmic scale for y-axis to see all curves clearly

plt.xlabel("n")
plt.ylabel("Time Complexity (log scale)")
plt.title("Running Time Comparisons")
plt.legend()
plt.grid(True)
plt.show()
````
